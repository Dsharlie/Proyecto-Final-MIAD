```python
import pandas as pd
import numpy as np
```

```python
# información dada
data = pd.ExcelFile("datos_contugas.xlsx")

#Como la información está distribuida por hoja, para cada cliente, se desea juntar toda la información en una sola tabla
#lista para almacenar los datos de cada hojoa

all_data = []

# iteración para extraer la información de cada hoja como Dataframe y añadirlo en la lista
for sheet_name in data.sheet_names:
    df = pd.read_excel("datos_contugas.xlsx", sheet_name= sheet_name)
    df['Cliente'] = sheet_name
    all_data.append(df)
 
 #Juntar todo la información en un solo dataframe
combined_data = pd.concat(all_data, ignore_index=True)

combined_data['Cliente'] = combined_data['Cliente'].astype(str).str.strip()
combined_data
```

```python
#Conocer alguna información por cliente

#Definir cuartiles
def q1(x):
    return x.quantile(0.25)
def q2(x):
    return x.median()
def q3(x):
    return x.quantile(0.75)


registro_fecha = combined_data.groupby('Cliente', as_index=False).agg(
    registros=('Cliente', 'count'),
    fecha_min=('Fecha', 'min'), 
    fecha_max=('Fecha', 'max'))

registro_fecha


```

```python
#registros nulos por cliente

# Contar registros nulos agrupados por cliente
null_counts = combined_data.groupby('Cliente').apply(lambda x: x.isnull().sum())
null_counts
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Asegúrate de que la columna de cliente esté correctamente definida
clientes = combined_data['Cliente'].unique()
variables = ['Presion', 'Volumen', 'Temperatura']  # Variables clave

# Crear un DataFrame para almacenar resultados de los registros fuera de rango por cliente y variable
resultados = []

# Calcular límites y registros fuera de rango por cliente y variable
for cliente in clientes:
    cliente_data = combined_data[combined_data['Cliente'] == cliente]
    for variable in variables:
        Q1 = cliente_data[variable].quantile(0.25)
        Q3 = cliente_data[variable].quantile(0.75)
        IQR = Q3 - Q1
        lower_limit = Q1 - 1.5 * IQR
        upper_limit = Q3 + 1.5 * IQR
        
        fuera_de_rango = cliente_data[(cliente_data[variable] < lower_limit) | (cliente_data[variable] > upper_limit)]
        total_registros = len(cliente_data)

        resultados.append({
            'Cliente': cliente,
            'Variable': variable,
            'Total_Registros': total_registros,
            'Registros_Fuera_Rango': len(fuera_de_rango),
            'Porcentaje_Fuera_Rango': (len(fuera_de_rango) / total_registros) * 100
        })

# Convertir resultados a DataFrame
resultados_df = pd.DataFrame(resultados)

# Generar gráficos para cada variable
for variable in variables:
    plt.figure(figsize=(14, 8))
    subset = resultados_df[resultados_df['Variable'] == variable]
    sns.barplot(data=subset, x='Cliente', y='Porcentaje_Fuera_Rango', palette='viridis')
    plt.title(f'Porcentaje de Registros Fuera de Rango - {variable}')
    plt.xlabel('Cliente')
    plt.ylabel('Porcentaje Fuera de Rango (%)')
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
```

```python
# Definir las variables clave
variables = ['Presion', 'Volumen', 'Temperatura']  # Ajusta según tus columnas
clientes = combined_data['Cliente'].unique()  # Columna que identifica a los clientes

# Crear un diccionario para almacenar los resultados
resultados = []

# Procesar cada cliente
total_registros_general = len(combined_data)  # Total general de registros

for cliente in clientes:
    cliente_data = combined_data[combined_data['Cliente'] == cliente]  # Filtrar por cliente
    total_registros_cliente = len(cliente_data)  # Total de registros por cliente
    
    for variable in variables:
        # Calcular estadísticas básicas
        Q1 = cliente_data[variable].quantile(0.25)
        Q3 = cliente_data[variable].quantile(0.75)
        IQR = Q3 - Q1
        lower_limit = Q1 - 1.5 * IQR
        upper_limit = Q3 + 1.5 * IQR

        # Contar registros fuera de rango
        fuera_de_rango = cliente_data[(cliente_data[variable] < lower_limit) | (cliente_data[variable] > upper_limit)]
        total_fuera_de_rango = len(fuera_de_rango)

        # Guardar resultados
        resultados.append({
            'Cliente': cliente,
            'Variable': variable,
            'Mínimo': cliente_data[variable].min(),
            'Máximo': cliente_data[variable].max(),
            'Q1': Q1,
            'Q3': Q3,
            'Límite Inferior': lower_limit,
            'Límite Superior': upper_limit,
            'Registros Totales Cliente': total_registros_cliente,
            'Registros Fuera de Rango': total_fuera_de_rango,
            'Porcentaje Fuera de Rango Cliente (%)': (total_fuera_de_rango / total_registros_cliente) * 100,
            'Porcentaje Fuera de Rango Total (%)': (total_fuera_de_rango / total_registros_general) * 100
        })

# Convertir resultados a un DataFrame
resultados_df = pd.DataFrame(resultados)

# Calcular totales generales por variable
totales_generales = resultados_df.groupby('Variable').agg({
    'Registros Fuera de Rango': 'sum',
    'Registros Totales Cliente': 'sum'
}).reset_index()

# Calcular el porcentaje total general
totales_generales['Porcentaje Total Fuera de Rango (%)'] = (
    totales_generales['Registros Fuera de Rango'] / total_registros_general
) * 100

resultados_df
totales_generales
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Crear un diccionario para almacenar la distribución de registros fuera de rango
distribucion = []

# Procesar cada cliente y variable
for cliente in clientes:
    cliente_data = combined_data[combined_data['Cliente'] == cliente]  # Filtrar por cliente
    
    for variable in variables:
        # Calcular estadísticas básicas
        Q1 = cliente_data[variable].quantile(0.25)
        Q3 = cliente_data[variable].quantile(0.75)
        IQR = Q3 - Q1
        lower_limit = Q1 - 1.5 * IQR
        upper_limit = Q3 + 1.5 * IQR

        # Contar registros fuera de rango
        fuera_de_rango = cliente_data[(cliente_data[variable] < lower_limit) | (cliente_data[variable] > upper_limit)]

        # Agregar distribución al resultado
        distribucion.append({
            'Cliente': cliente,
            'Variable': variable,
            'Registros Fuera de Rango': len(fuera_de_rango),
            'Porcentaje Fuera de Rango del Cliente (%)': (len(fuera_de_rango) / len(cliente_data)) * 100
        })

# Convertir la distribución a un DataFrame
distribucion_df = pd.DataFrame(distribucion)

# Crear gráficos de distribución con seaborn
# Gráfico 1: Barras por Cliente y Variable mostrando registros fuera de rango
plt.figure(figsize=(12, 8))
sns.barplot(data=distribucion_df, x='Cliente', y='Registros Fuera de Rango', hue='Variable', ci=None)
plt.title('Registros Fuera de Rango por Cliente y Variable')
plt.xlabel('Cliente')
plt.ylabel('Registros Fuera de Rango')
plt.xticks(rotation=45)
plt.legend(title='Variable')
plt.tight_layout()
plt.show()

# Gráfico 2: Barras por Cliente y Variable mostrando el porcentaje de registros fuera de rango
plt.figure(figsize=(12, 8))
sns.barplot(data=distribucion_df, x='Cliente', y='Porcentaje Fuera de Rango del Cliente (%)', hue='Variable', ci=None)
plt.title('Porcentaje de Registros Fuera de Rango por Cliente y Variable')
plt.xlabel('Cliente')
plt.ylabel('Porcentaje Fuera de Rango (%)')
plt.xticks(rotation=45)
plt.legend(title='Variable')
plt.tight_layout()
plt.show()
```
